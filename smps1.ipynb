```python
import pandas as pd

# Load the dataset
file_path = '/home/coliewo/Desktop/SMPS_data/first_trial.csv'
df = pd.read_csv(file_path)

# Display the first few rows and columns
df.head(), df.columns
```

```python
import numpy as np
import xarray as xr

# Convert the 'Time' column to datetime
df['Time'] = pd.to_datetime(df['Time'], format='%d/%m/%Y %H:%M:%S')

# Set 'Time' as the index
df.set_index('Time', inplace=True)
```

```python

# Separate the columns into positive and negative particles
positive_columns = [col for col in df.columns if not col.startswith('_')]
negative_columns = [col for col in df.columns if col.startswith('_')]
```

```python

# Extract the diameters and convert to float
positive_diameters = np.array([float(col) for col in positive_columns])
negative_diameters = np.array([float(col[1:]) for col in negative_columns])  # Remove the underscore
```

```python

# Resample the data to hourly intervals by averaging
df_hourly = df.resample('H').mean()
```

```python
df_hourly
```

```python

# Calculate dN/dlogDp
def calculate_dn_dlogdp(df, diameters):
    log_diameters = np.log10(diameters)
    dlogdp = np.diff(log_diameters)
    dN_dlogDp = df.iloc[:, 1:].values / dlogdp
    return dN_dlogDp

# Apply the calculation for positive and negative particles
dN_dlogDp_positive = calculate_dn_dlogdp(df_hourly[positive_columns], positive_diameters)
dN_dlogDp_negative = calculate_dn_dlogdp(df_hourly[negative_columns], negative_diameters)
```

```python
diam = positive_diameters[:-1]
```

```python
df_hourly.index
```

```python
ds = xr.Dataset(
    data_vars=dict(
        positive_particles=(["time", "diameter"], dN_dlogDp_positive),
        negative_particles=(["time", "diameter"], dN_dlogDp_negative),
    ),
    coords=dict(
        date=df_hourly.index,
        diameter=diam,
        reference_time = pd.Timestamp("2024-06-24"),
    ),
    attrs=dict(description="SMPS data."),
)
ds
```

```python
def calculate_dn_dlogdp(df, diameters):
    log_diameters = np.log10(diameters)
    dlogdp = np.diff(log_diameters)
    dN_dlogDp = df.iloc[:, 1:].values / dlogdp[:, np.newaxis]
    return dN_dlogDp, diameters[:-1]  # Exclude the last diameter to match the size

# Apply the calculation for positive and negative particles
dN_dlogDp_positive, positive_diameters_adjusted = calculate_dn_dlogdp(df_hourly[positive_columns], positive_diameters)
dN_dlogDp_negative, negative_diameters_adjusted = calculate_dn_dlogdp(df_hourly[negative_columns], negative_diameters)

# Create an xarray dataset
ds = xr.Dataset(
    {
        'positive_particles': (('date', 'diameters'), dN_dlogDp_positive.T),
        'negative_particles': (('date', 'diameters'), dN_dlogDp_negative.T),
    },
    coords={
        'date': df_hourly.index,
        'diameters': positive_diameters_adjusted  # Using adjusted diameters to match data size
    }
)

ds
```
